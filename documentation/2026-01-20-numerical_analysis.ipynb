## Observations
1. **BLEU** scores demonstrate limited utility for short-form responses due to insufficient n-gram overlap at higher-order (n=4) sequences.
2. This limitation affects **ROUGE-2** metrics for the same reason.
- **To do:**
- Implement length-aware weighting schemes that proportionally reduce the influence of n-gram metrics for responses below optimal length thresholds.
- Compare llm-answers with a series of references that contain higher lexical diversity
- Add lemmatization, synonyms or related lexica to find word matches
- Accuracy feedbacks should take into consideration, that accuracy measure may not be adecuate for questions of type "sensitive" or "creative", since the possibilities for llm-answers are more diverse for these types of questions. We would need more reference-answers and more diverse accuracy-measures (including synonyms, lemmatization, and related word lexica)
3. **Exact match** measures are too strict and yield zero for all question/answer pairs
- **To do:**
- Decrease this measure's weight when calculating the **composite accuracy** 
4. **Semantic relevance** yields particularly good results
- **To do:**
- Weight this measure higher on **composite accuracy** computation
5. **Numeric accuracy** is high for question/answer-pairs that do not contain any numbers. Thus, the overall score is artificially increased
- **To do:**
- Only use this measure to calculate the overall score in question/llm-answer pairs that contain numbers of contextual significance
- Exclude this measure for "instruction" questions, since these contain numbered steps that should not be evaluated by this measure
6. **Content coverage** yields satistfying results. Lemmatization and synonyme coverage seems to be an effective tool
- **To do:** Increase this measure' weight for calculation of overall score
7. **Feedback:** Feedbacks are not properly reflect how accurately references are reproduced by llm-answers
- **To do:**
- Adapt thresholds: Since some measures are not well adapted (according to the mentioned points above), many **composite accuracy** measures are too low. Therefore, lower thresholds for accuracy feedbacks
- Accuracy measures are not well adapted to "creative" and "sensitive" questions", since they offer a wider variety of possible llm-answers that can vary from references. This can be countered by including different references, which additionally contain a wider range of keywords
8. **Composite accuracy:** In light of the accuracies previously mentioned, the composite accuracy does not properly reflect the relationship between references and llm-answers.
- **To do:**
- Correct how each meausre contributes to the composite accuracy-score. Less suitable measure should be downweighed, while more accurate measures upweighed.
- "Creative" and "sensitive" questions should have seperate weights, taking into consideration that they are less "accurate" by nature.
9. **Passed_accuracy**: Measure is too strict for "creative" and "sensitive" questions, since these allow for higher response freedom.
- **To do:**
- Adapt thresholds for creative" and "sensitive" questions. The meaure is not appropriate for these categories.
- Question with "id" 12 is wrongly categorized as false. Correct the remaining measures' inadequacies to make sure the question passes the threshold.

# 2. Relevance features

````python
# Set display to show ALL rows and columns
pd.set_option('display.max_rows', None)    # Show all rows
pd.set_option('display.max_columns', None) # Show all columns
pd.set_option('display.width', None)       # Auto-detect terminal width
pd.set_option('display.max_colwidth', None) # Show full column content

'''1. sort by category'''
for cat in ["Factual", "Explanatory", "Creative", "Instruction", "Sensitive"]:
    df1 = df[df["category"] == cat]
    #display(df1)    
    '''2. Search for relevance features'''
    rel_features = [feat for feat in df1.columns if "relevance" in feat]
    metadata = ["category", "question", "reference", "response"]
    cols = metadata + rel_features
    df2 = df1[cols]
    #print(df2.columns)
    #display(df2)
    '''3. search for specific category'''
    keywords=["composite", "feedback","passed"]
    feature = [feat for feat in df2.columns if any(keyw in feat for keyw in keywords)]
    cols = metadata + feature
    df3 = df2[cols]
    #display(df3)

    df4 = df3[df3["passed_relevance"]==False]
    display(df4)
````

## Observations

1. **Semantic relevance** yields particularly good results
- **To do:**
- Weight this measure higher on overall-score computation
2. **Term frequency-inverse document frequency** approaches (tfidf) demonstrate reduced effectiveness when used with small text corpora. The method involves calculating cosine similarities between question- and llm-answer vectors. Words found in both - question and llm-answer - share components, thus increasing the cosine similiarity. Vector-components standing for words with a high appearence across documents are downsized while word with a rare appearance across text corpora are upsized. Words that are not found in the text corpus are simply ignored and set to zero. Since the used text-corpus is too small, this happens to most components, such that most ft-idf scores are zero
- **To do:**
- Increase document corpus by calling outside libraries
- Add questions and reference answers to text corpus. This method creates a domain-specific corpus, avoiding that a word is not present in the corpus. However, it may lead to overfitting or a skewed IDF. Thus, it should only be applied in combination with an increased corpus (point above)
- Reduce the weights for this measure when calculating the overall score.
3. **keyword overlap** yields satistfying results. Lemmatization and synonyme coverage seems to be an effective tool
- **To do:** Increase this measure' weight for calculation of overall score
4. **Intent-match** does not find any matches
- **To do:** Increase vacabulary to be searched for
5. **Refusal-score** successfully found any refusal in llm-answers
6. **Depth-score** category specificity works approproately.
- **To do:**
- Add functionality to customize minimal and maximal resonse lengths
7. **Composite-relevance:** Values are generally too low even for appropriately tuned llm-answers to the question.
 - **To do:**
 - Fine-tune weights of different measures as described in the previous points.
8. **Relevance feedback:** Feedback is too strict, even for llm-answers well tuned to the respective questions
 - **To do:**
 - Correct **composite-relevance**-measures to obtain higher scores
 - Lower thresholds if relevance measures cannot be improved yet
9. **Passed Relevance**: This measure is reflecting the same difficulties encoutered for the **composite relevace** measure.
 - **To do:**
 -  Correct **composite-relevance**-measures to obtain higher scores
 -  Lower thresholds if relevance measures cannot be improved yet

10. **"creative" and "sensitive" questions**:
As already discussed in section ```1. Accuracy features```, this question categories allow for more freedom in llm-responses. Thus, their accuracy and relevance measures are unrealistically low.
- **To-do:**
- Adjust weights for these catgories specifically. Semantic relevance seems to be the most appropriate measure.
- Expand reference answers by adding more diverse keywords and comparing the llm-responses to a diversity of different references.
- Add features that make use of synonyms and lemmatization in the other measures

## Observations
1. **Length:** Measure is appropriate for the test-data
**To-do:**
- Adapt minimum and maximum length thresholds to the question-length and category. "Creative" and "sensitive" categories may yield longer llm-answers
2. **Fluency-score:** Measure is appropriate for the test-data
**To-do**
-At current, this measure only includes Capitalization and word-lengths. One could add algorithms to check further grammatical use and sentence complexity-
3. **Coherence-score:** Measure is inappropiate for certain categories and the normalization factor is set too strcitly. The factor assumes one coherence indicator per sentence transition, which is too high.
**To-do:**
- Reduce the normalization factor.
- Adapt the normalization factor to each catgeory. "Factual" and "Instrucution" questions are less likely to use coherence indicators.
4. **Conciseness-score:** Conciseness-scores generally high.
5. **Readability-score:** This measure uses the Flesch Reading Ease (FRE) formula, however, all results yield zero. The factor $\frac{0.39 \cdot WPS + 11.8 \cdot AWL - 15.59}{20}$ in the equation $ R_d = \max(0, \min(1, 1 - \frac{0.39 \cdot WPS + 11.8 \cdot AWL - 15.59}{20})) $ (where $ WPS $ = words per sentence, $ AWL $ = average word length) is therefore too small and needs to be adapted.
- **To-do:**
- Adapt the Flesch Reading Ease (FRE) formula to more concise llm-answers
6. **Composite-quality** is heavily influenced by zeros in **conciseness** and **readibility scores**.
**To-do:**
  - Correct readability and conciseness scores to correct the composite-quality.  
7. **Length-feedback**: Appropiate
8. **Quality-feedback:** This measure uses **composite-quality**-values as a reference. Since these values are too low for some llm-anwers, the feedback is also inaccurate
**To-do:**
  - Correct readability and conciseness scores to correct the composite-quality.
9. **Passed quality:** This measure uses **composite-quality**-values as a reference. Since these values are too low for some llm-anwers, they do not pass the requiered threshold.
**To-do:**
- Correct readability and conciseness scores to correct the composite-quality.
- Adapt thresholds to que question's catgegory. "Factual" and "Instrucution" will naturally showcase lower quality scores.
